```python
import numpy as np
import pandas as pd
```

```python
from numpy.linalg import svd
from sklearn.decomposition import TruncatedSVD,PCA
import matplotlib.pyplot as plot
```

```python
pca = PCA(2)
t_svd = TruncatedSVD(2)
```

```python
size = 100
```

Generate Dataset

```python
data = {'Loc':np.random.normal(80,20,size),'Size':np.random.normal(80,20,size),'Mall_Near':np.random.normal(80,20,size),'Ambience':np.random.normal(80,20,size),'Rent':np.random.normal(80,20,size)}
```

```python
data = pd.DataFrame(data)
```

```python
data
```

Drop Target

```python
features=data.drop('Rent',axis=1)
```

```python
features
```

Dataframe to numpy

```python
features_array = features.to_numpy()
```

```python
features_array
```

```python
t_svd.fit_transform(features_array)
```

singular

```python
pca.fit_transform(features_array)
```

```python
plot.plot(pca.explained_variance_ratio_)
plot.ylabel('Explained Variance')
plot.xlabel('Components')
plot.show()
```
